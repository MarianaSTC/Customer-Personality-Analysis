```python
df['Income'].fillna(df['Income'].median(), inplace=True)

df = df.drop(['Z_CostContact', 'Z_Revenue', 'ID'], axis=1)

df['Age'] = 2024 - df['Year_Birth']
df['NbYears_customer_enrollment'] = np.round((pd.to_datetime('01/01/2024') - pd.to_datetime(df['Dt_Customer'], dayfirst=True)).dt.days / 365, 1)
df = df.drop(['Year_Birth', 'Dt_Customer'], axis=1)

df.drop_duplicates(inplace=True)

df["Spent"] = df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"]
df["Children"] = df["Kidhome"] + df["Teenhome"]
df["Living_With"] = df["Marital_Status"].replace(
    {"Married":1, "Together":1, "Absurd":0, "Widow":0, "YOLO":0, "Divorced":0, "Single":0, "Alone":0}
    )
df["Education"] = df["Education"].replace(
    {"Basic":0,"2n Cycle":0, "Graduation":1, "Master":2, "PhD":2}
    )
df.drop(['Marital_Status', 'Kidhome', 'Teenhome'], axis=1, inplace=True)

df.drop(['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
       'AcceptedCmp2', 'Complain'], axis=1, inplace=True)

df_clean = df[df['Income']<600000]
df_clean = df_clean[df_clean['Age']<110]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter
```

```python
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ])

categorical_transformer = Pipeline(steps=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, make_column_selector(dtype_include=np.number)),
    ('cat', categorical_transformer, make_column_selector(dtype_include=object))
    ])

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ])
```

```python
df_encoded = pipeline.fit_transform(df_clean)
```

```python
print(df_clean.shape)
print(df_encoded.shape)
```

```python
k_range = range(2, 30)

kmeans_per_k = [KMeans(n_clusters=k, n_init='auto').fit(df_encoded) for k in k_range]
inertias = [model.inertia_ for model in kmeans_per_k]

fig = plt.figure(figsize=(10, 4))
plt.scatter(k_range, inertias)
plt.plot(k_range, inertias)
plt.grid(True)
plt.title('Elbow curve')
plt.ylabel('Inertie')
plt.xlabel('Nombre de clusters')
plt.show()
```

```python
k_range = range(2, 30)
n = 20
inertias = []

for k in k_range:
  inertias.append(np.mean([KMeans(n_clusters=k, n_init='auto').fit(df_encoded).inertia_ for i in range(n)]))
```

```python
fig = plt.figure(figsize=(10, 4))
plt.scatter(k_range, inertias)
plt.plot(k_range, inertias)
plt.grid(True)
plt.title('Elbow curve')
plt.ylabel('Inertie moyenne')
plt.xlabel('Nombre de clusters')
plt.show()
```

```python
k_range = range(2, 21)

kmeans_per_k = [KMeans(n_clusters=k, n_init='auto').fit(df_encoded) for k in k_range]

silhouette_scores  = [silhouette_score(df_encoded, model.labels_) for model in kmeans_per_k]

fig = plt.figure(figsize=(10, 4))
plt.scatter(k_range, silhouette_scores)
plt.plot(k_range, silhouette_scores)
plt.grid(True)
plt.title('Silhouette score')
plt.xlabel('Nombre de clusters')
plt.show()
```

```python
k_range = range(2, 21)
n = 20
silhouette_scores = []

for k in k_range:
  silhouette_scores.append(np.mean([silhouette_score(df_encoded, KMeans(n_clusters=k, n_init='auto').fit(df_encoded).labels_) for i in range(n)]))
```

```python
fig = plt.figure(figsize=(10, 4))
plt.scatter(k_range, silhouette_scores)
plt.plot(k_range, silhouette_scores)
plt.grid(True)
plt.title('Moyenne silhouette score')
plt.xlabel('Nombre de clusters')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

```python
db = DBSCAN().fit(df_encoded)

labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)
```

```python
db.labels_
```

```python
eps = [0.1, 0.3, 0.5, 0.8, 1, 3, 5, 10, 30, 50, 100]

min_samples = [2, 5, 10, 25, 50]

df_results = pd.DataFrame()

for epsilon in eps:
  for min_samp in min_samples:
    df_row = pd.DataFrame(columns=['eps', 'min_samples', 'n_clusters', 'n_noise'], index=[0])
    df_row['eps'] = epsilon
    df_row['min_samples'] = min_samp
    db = DBSCAN(eps=epsilon, min_samples=min_samp).fit(df_encoded)
    labels = db.labels_
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise_ = list(labels).count(-1)
    df_row['n_clusters'] = n_clusters_
    df_row['n_noise'] = n_noise_
    df_results = pd.concat([df_results, df_row], axis=0)
```
